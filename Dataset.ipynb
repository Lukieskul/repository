```python
path1="https://raw.githubusercontent.com/Lukieskul/repository/main/WHR20_DataForFigure2.1.csv"
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df=pd.read_csv(path1)
df.head()
```

```python
df.columns
```

```python
plt.scatter(df['Logged GDP per capita'], df['Healthy life expectancy'])
plt.xlabel('Logged GDP per capita')
plt.ylabel('Healthy life expectancy')
plt.title('Relacion entre GDP y expectativa de vida')
plt.show()
```

```python
plt.scatter(df['Logged GDP per capita'], df['Healthy life expectancy'])
plt.xlabel('Logged GDP per capita')
plt.ylabel('Healthy life expectancy')
plt.title('Relacion entre GDP y expectativa de vida')
plt.show()
```

```python
df.groupby('Regional indicator')['Ladder score'].sum()
```

```python
import matplotlib.pyplot as plt
df_grouped = df.groupby('Regional indicator')['Ladder score'].sum().sort_values(ascending=False)
df_grouped.plot(kind='bar')
plt.xlabel('Regional indicator')
plt.ylabel('Sum of Ladder score')
plt.title('Sum of Ladder Scores by Regional Indicator')
plt.show()
```

```python
plt.scatter(df['Logged GDP per capita'], df['Ladder score'])
plt.xlabel('Logged GDP per capita')
plt.ylabel('Ladder score')
plt.title('Relacion entre  GDP felicidad en general')
plt.show()
```
